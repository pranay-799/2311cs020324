```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from scipy.stats import zscore

data=sns.load_dataset('iris')
df=data.copy()
```

```python
df
```

```python
z_scores = np.abs(zscore(df.drop('species',axis=1)))
z_scores
```

```python
zscore(df.drop('species',axis=1))
```

```python
non_outliers = (z_scores <= 3).all(axis=1)
df_no_outliers = df[non_outliers]
```

```python
df_no_outliers
```

```python
(z_scores<3).all(axis=1)
```

```python
outliers = (z_scores > 3).any(axis=1)

outlier_rows = df[outliers]
outlier_rows
```

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=sns.load_dataset('iris')
data
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data['species']=lb.fit_transform(data['species'])
```

```python
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data)
```

```python
clf.predict(data)
```

```python
y_hat=clf.predict(data)
y_hat
```
